### 1. Analyzing average stock-wise performance over the years (2006-2018)

##### As we can see fromt the chart, stocks with high average volume were (AAPL, CSCO, GE, INTC, MSFT)

```python
df = allstock[["Volume","Name"]].groupby(["Name"]).agg({'Volume':['mean']})
fig=df.plot(kind='bar',fontsize=24,figsize=(30,10),legend=True)
fig.legend(loc=1,fontsize=24)
plt.xlabel('Name of the stock',fontsize=24)
plt.ylabel('Avg Volume',fontsize=24)
plt.title('Average Stock Performance 2006-2018',fontsize=24)
plt.show()
```

### 2. Now we will deep dive into this stocks with higher average volume and analyse the performance

#### Now let's load the individual stock data into the dataframes

```python
aapl_data = pd.read_csv('AAPL_2006-01-01_to_2018-01-01.csv')
csco_data = pd.read_csv('CSCO_2006-01-01_to_2018-01-01.csv')
ge_data = pd.read_csv('GE_2006-01-01_to_2018-01-01.csv')
intc_data = pd.read_csv('INTC_2006-01-01_to_2018-01-01.csv')
msft_data = pd.read_csv('MSFT_2006-01-01_to_2018-01-01.csv')
```

```python
aapl_stock_date = aapl_data['Date']
aapl_stock_close = aapl_data['Close']
aapl_stock_date = pd.to_datetime(aapl_stock_date,infer_datetime_format=True)
aapl_frames = [aapl_stock_date, aapl_stock_close]
aapl = pd.concat(aapl_frames, axis=1)
aapl = aapl.set_index(['Date'])
aapl = aapl.sort_index()

csco_stock_date = csco_data['Date']
csco_stock_close = csco_data['Close']
csco_stock_date = pd.to_datetime(csco_stock_date,infer_datetime_format=True)
csco_frames = [csco_stock_date, csco_stock_close]
csco = pd.concat(csco_frames, axis=1)
csco = csco.set_index(['Date'])
csco = csco.sort_index()

ge_stock_date = ge_data['Date']
ge_stock_close = ge_data['Close']
ge_stock_date = pd.to_datetime(ge_stock_date,infer_datetime_format=True)
ge_frames = [ge_stock_date, ge_stock_close]
ge = pd.concat(ge_frames, axis=1)
ge = ge.set_index(['Date'])
ge = ge.sort_index()

intc_stock_date = intc_data['Date']
intc_stock_close = intc_data['Close']
intc_stock_date = pd.to_datetime(intc_stock_date,infer_datetime_format=True)
intc_frames = [intc_stock_date, intc_stock_close]
intc = pd.concat(intc_frames, axis=1)
intc = intc.set_index(['Date'])
intc = intc.sort_index()

msft_stock_date = msft_data['Date']
msft_stock_close = msft_data['Close']
msft_stock_date = pd.to_datetime(msft_stock_date,infer_datetime_format=True)
msft_frames = [msft_stock_date, msft_stock_close]
msft = pd.concat(msft_frames, axis=1)
msft = msft.set_index(['Date'])
msft = msft.sort_index()
```

```python
plt.plot(aapl, color='blue', label='AAPL')
plt.plot(csco, color='red', label='CSCO')
plt.plot(ge, color='orange', label='GE')
plt.plot(intc, color='green', label='INTC')
plt.plot(msft, color='yellow', label='MSFT')
plt.figsize=(50,15)
plt.xlabel('Year',fontsize=12)
plt.ylabel(' Stock Closing Value',fontsize=12)
plt.legend(loc='upper left')
plt.show()
```

### 4 and 5. Just out of curiousity, let's check if Apple and Microsoft volume were correlated

```python
x=aapl_data[['Volume']]
y=msft_data['Volume']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.333, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
# print the intercept
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,x.columns,columns=['Coefficient'])
coeff_df
```

```python
predictions = lm.predict(x_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.distplot((y_test-predictions),bins=50);
```
